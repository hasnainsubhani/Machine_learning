```python
l=[]
for i in dt.columns:
    for j in range(dt.shape[0]):
        if dt[i][j]==' ?':
            l.append(j)
dt.drop(index=l,inplace=True)
```

```python
dt.groupby(['workclass'])['workclass'].count()
```

```python
dt.groupby(['education'])['education'].count()
```

```python
#cleaning unwanted space in object columns
for i in dt.columns:
    if dt[i].dtype=='object':
        dt[i] = dt[i].str.strip()
```

```python
dt.head()
```

```python
dt['person_makes']=dt['person_makes'].map({'<=50K':0,'>50K':1})
```

# Sex wise earnings

```python
 dt_copy.groupby(['person_makes','sex'])['person_makes'].count()
```

```python
sex_wise_earning = pd.crosstab(dt_copy['person_makes'],dt_copy['sex'])
sex_wise_earning.plot(kind='bar')
```

# Age less than or equall to 30 and earns more than 50k

```python
dt_less_than_equall_to_30 = dt[dt['age']<=30]
```

```python
#No of people whose age above 30 earn more than 50K

dt_less_than_equall_to_30[dt_less_than_equall_to_30['person_makes']==1]
```

```python
dt_ageupto_30_earns_morethan50k = dt_less_than_equall_to_30[dt_less_than_equall_to_30['person_makes']==1]
```

```python
sex_wise_earning = pd.crosstab(dt_ageupto_30_earns_morethan50k['person_makes'],dt_ageupto_30_earns_morethan50k['sex'])
sex_wise_earning.plot(kind='bar')
```
